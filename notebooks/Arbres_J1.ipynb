```python
#calcul du nombre d'arbres plantés en moyenne :
tot_arbres = len(lines)-1 # on prend tous les arbres bien qu'on ne connaisse pas toujours leur date
annees = [int(elem) for elem in anneePlantation[1:] if elem!=''] # conversion du type de la liste
# et suppression des valeurs manquantes
debut = min(annees)
fin = max(annees)
print("Entre {} et {}, il y a eu {} arbres de plantés.".format(debut,fin,tot_arbres))
print("Soit {:.2f} arbres par an en moyenne".format(float(tot_arbres/(fin-debut))))
```

```python
#affichage du nombre d'arbre depuis l'année de notre naissance
anneeNaissance = 1991
count = 0
for annee in annees : # travail sur la liste numerique créee plus haut
    if annee > anneeNaissance :
        count+=1
print("%d arbres ont été plantés depuis %d." % (count,anneeNaissance))
```

```python
def nb_arbres_entre(list,annee1,annee2) :
    """Renvoie  le nombre d'elements de liste compris entre 
    annee1 (inclus) et annee2 (exclus)"""
    count = 0
    for elem in list :
        if elem>=annee1 and elem<annee2 :
            count+=1
    return(count)
```

```python
maire_mandat = [
    {'nom':'Mistral', 'deb':1919, 'fin':1932},
    {'nom':'Martin', 'deb':1932, 'fin':1935},
    {'nom':'Cocat', 'deb':1932, 'fin':1944},
    {'nom':'Lafleur', 'deb':1944, 'fin':1945},
    {'nom':'Martin', 'deb':1945, 'fin':1947},
    {'nom':'Bally', 'deb':1947, 'fin':1948},
    {'nom':'Perinetti', 'deb':1948, 'fin':1949},
    {'nom':'Martin', 'deb':1949, 'fin':1959},
    {'nom':'Michallon', 'deb':1959, 'fin':1965},
    {'nom':'Dubedout', 'deb':1965, 'fin':1983},
    {'nom':'Carignon', 'deb':1983, 'fin':1995},
    {'nom':'Destot', 'deb':1995, 'fin':2014},
    {'nom':'Piolle', 'deb':2014, 'fin':2020}
    ] # liste des maires de Grenoble depuis 1919 avec les années de début et de fin de mandat
```

```python
# nombre d'arbres par maire
res = {}
for maire in maire_mandat :
    if maire['nom'] in res :
        res[maire['nom']] += nb_arbres_entre(annees,maire['deb'],maire['fin'])
    else :
        res[maire['nom']] = nb_arbres_entre(annees,maire['deb'],maire['fin'])
arbres_max = 0
nom_maire = ''
for maire in res :
    if res[maire] > arbres_max :
        arbres_max = res[maire]
        nom_maire = maire
#print(res)
print("Le maire qui a le plus planté d'arbres est %s avec %d arbres." % (nom_maire,arbres_max))
# autre syntaxe pour afficher des variables
```

Comparaison avec le fichier corrompu

```python
corrupted_data_path = data_path.parent / "ESP_PUBLIC.IDENTITE_ARBRE_CORRPUTED.csv"
print(corrupted_data_path)
```

```python
# Comparaison directe des deux fichiers et affichage des différences
with open(corrupted_data_path) as corr_data :
    with open(data_path) as data :
        corr_csvfile = csv.reader(corr_data)
        csvfile = csv.reader(data)
        for (lineCorr, line) in zip(corr_csvfile,csvfile) :
            if lineCorr!=line :
                for (elemCorr,elem) in zip(lineCorr,line) :
                    if elemCorr!=elem :
                        print(elemCorr,elem)
```

```python
# Comparaison directe des deux fichiers et stockage de la bonne version
with open(corrupted_data_path) as corr_data :
    with open(data_path) as data :
        new_lines = []
        corr_csvfile = csv.reader(corr_data)
        csvfile = csv.reader(data)
        for (lineCorr, line) in zip(corr_csvfile,csvfile) :
            new_lines.append(lineCorr)
            if lineCorr!=line :
                pos = 0
                for (elemCorr,elem) in zip(lineCorr,line) :
                    if elemCorr!=elem :
                        print(elemCorr,elem)
                        #print(new_lines[-1])
                        new_lines[-1][pos] = elem
                    pos+=1
                
updated_data_path = data_path.parent / "ESP_PUBLIC.IDENTITE_ARBRE_UPDATED.csv"
with open(updated_data_path,"w",newline="") as updated_data :
    csv_writer = csv.writer(updated_data)
    csv_writer.writerows(new_lines) # pour écrire plusieures lignes d'un coup, 
    # writerow() pour écrire une seule ligne
```

```python
# Comparaison de la nouvelle version et de la référence
with open(updated_data_path) as updated_data :
    with open(data_path) as data :
        up_csvfile = csv.reader(updated_data)
        csvfile = csv.reader(data)
        erreur = False
        for (lineUp, line) in zip(up_csvfile,csvfile) :
            if lineUp!=line :
                print('Erreur ! Revoie ton code')
                erreur = True
                for (elemUp,elem) in zip(lineUp,line) :
                    if elemUp!=elem :
                        print(f"{elemUp} != {elem}")
if not erreur :
    print('Bravo, ta mission est accomplie.')
```
